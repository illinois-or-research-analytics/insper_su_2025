```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "font.family": "Helvetica",
    "font.size": 16
})

# Load and clean
py_file = "data/experiments/python/6_23_ecc7_fmf_no_ss_ra_pubmed_python_abm/output/output.aux"
cpp_file = "data/experiments/cpp/6_18_ecc7_fmf_standard_no_ss_ra_pubmed/output/output.aux"
df_py = pd.read_csv(py_file)
df_cpp = pd.read_csv(cpp_file)

df_py['in_degree'] = pd.to_numeric(df_py['in_degree'], errors='coerce')
df_cpp['in_degree'] = pd.to_numeric(df_cpp['in_degree'], errors='coerce')

df_py = df_py[df_py['in_degree'].notnull() & (df_py['in_degree'] >= 1)]
df_cpp = df_cpp[df_cpp['in_degree'].notnull() & (df_cpp['in_degree'] >= 1)]

# Restrict to top 5%
py_sorted = df_py.sort_values('in_degree', ascending=False)
cpp_sorted = df_cpp.sort_values('in_degree', ascending=False)

py_top5 = py_sorted.iloc[:int(0.05 * len(py_sorted))]
cpp_top5 = cpp_sorted.iloc[:int(0.05 * len(cpp_sorted))]

# Compute frequencies just for top 5%
py_counts = py_top5['in_degree'].value_counts().sort_index()
cpp_counts = cpp_top5['in_degree'].value_counts().sort_index()

py_deg = py_counts.index.values
py_freq = py_counts.values

cpp_deg = cpp_counts.index.values
cpp_freq = cpp_counts.values

# Plot on log-log
plt.figure(figsize=(10,6))
plt.scatter(np.log10(py_deg), py_freq, alpha=0.7, label="Python")
plt.scatter(np.log10(cpp_deg), cpp_freq, alpha=0.7, label="C++")

plt.xlabel("log10(in_degree)")
plt.ylabel("log10(frequency)")
plt.title("Degree distribution (Top 5% Python vs C++) NO SS RA SJ")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.style.use('ggplot')


py_file = "data/experiments/python/6_23_ecc7_fmf_no_ss_ra_pubmed_python_abm/output/output.aux"
cpp_file = "data/experiments/cpp/6_18_ecc7_fmf_standard_no_ss_ra_pubmed/output/output.aux"
df_py = pd.read_csv(py_file)
df_cpp = pd.read_csv(cpp_file)

df_py['in_degree'] = pd.to_numeric(df_py['in_degree'], errors='coerce')
df_cpp['in_degree'] = pd.to_numeric(df_cpp['in_degree'], errors='coerce')

df_py = df_py[df_py['in_degree'].notnull() & (df_py['in_degree'] >= 1)]
df_cpp = df_cpp[df_cpp['in_degree'].notnull() & (df_cpp['in_degree'] >= 1)]

plt.figure(figsize=(10,6))

# Manually assign colors
colors = {"Python": "steelblue", "C++": "darkorange"}

for df, version in [(df_py, "Python"), (df_cpp, "C++")]:
    for typ in ["agent", "seed"]:
        subset = df[df['type'] == typ]
        if subset.empty:
            continue
        counts = subset['in_degree'].value_counts().sort_index()
        deg = counts.index.values
        freq = counts.values
        linestyle = "--" if typ == "agent" else "-"
        plt.plot(np.log10(deg), freq, linestyle=linestyle, color=colors[version], label=f"{version} {typ}")

plt.xlabel("log10(in_degree)", fontname="Helvetica", fontsize=16)
plt.ylabel("frequency", fontname="Helvetica", fontsize=16)
plt.title("In-degree distribution by type (Python vs C++) NO SS RA", fontname="Helvetica", fontsize=16)

plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams.update({
    "font.family": "Helvetica",
    "font.size": 16
})

# ------------------------
# List your experiments
# ------------------------
experiments = [
    (
        ("data/experiments/python/6_23_ecc7_fmf_no_ss_ra_pubmed_python_abm/output/output.aux",
         "data/experiments/cpp/6_18_ecc7_fmf_standard_no_ss_ra_pubmed/output/output.aux"),
        "NO SS RA"
    ),
    (
        ("data/experiments/python/6_23_ecc7_fmf_ss_ra_pubmed_python_abm/output/output.aux",
         "data/experiments/cpp/6_18_ecc7_fmf_standard_ss_ra_pubmed/output/output.aux"),
        "SS RA"
    ),
    (
        ("data/experiments/python/6_23_ecc7_fmf_no_ss_sa_pubmed_python_abm/output/output.aux",
         "data/experiments/cpp/6_19_ecc7_fmf_standard_no_ss_sa_pubmed/output/output.aux"),
        "NO SS SA"
    ),
    (
        ("data/experiments/python/6_26_ecc7_fmf_ss_sa_pubmed_python_abm/output/output.aux",
         "data/experiments/cpp/6_19_ecc7_fmf_standard_ss_sa_pubmed/output/output.aux"),
        "SS SA"
    )
]

# ------------------------
# Set up quad panel
# ------------------------
fig, axes = plt.subplots(2, 2, figsize=(16,12))
axes = axes.flatten()

colors = {"Python": "steelblue", "C++": "darkorange"}

# ------------------------
# Loop over each experiment
# ------------------------
for ax, ((py_file, cpp_file), title) in zip(axes, experiments):
    # Load data
    df_py = pd.read_csv(py_file)
    df_cpp = pd.read_csv(cpp_file)
    
    # Clean data
    df_py['in_degree'] = pd.to_numeric(df_py['in_degree'], errors='coerce')
    df_cpp['in_degree'] = pd.to_numeric(df_cpp['in_degree'], errors='coerce')
    df_py = df_py[df_py['in_degree'].notnull() & (df_py['in_degree'] >= 1)]
    df_cpp = df_cpp[df_cpp['in_degree'].notnull() & (df_cpp['in_degree'] >= 1)]
    
    # Plot
    for df, version in [(df_py, "Python"), (df_cpp, "C++")]:
        for typ in ["agent", "seed"]:
            subset = df[df['type'] == typ]
            if subset.empty:
                continue
            counts = subset['in_degree'].value_counts().sort_index()
            deg = counts.index.values
            freq = counts.values
            linestyle = "--" if typ == "agent" else "-"
            ax.plot(np.log10(deg), freq, linestyle=linestyle, color=colors[version], label=f"{version} {typ}")
    
    ax.set_title(f"In-degree distribution {title}")
    ax.set_xlabel("log10(in_degree)")
    ax.set_ylabel("frequency")
    ax.grid(alpha=0.3)

# Avoid duplicate legends
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# All your experiments with custom tags
experiments = [
    ("cpp_noss_ra", "data/experiments/cpp/6_18_ecc7_fmf_standard_no_ss_ra_pubmed/output/output.aux"),
    ("py_noss_ra",  "data/experiments/python/6_23_ecc7_fmf_no_ss_ra_pubmed_python_abm/output/output.aux"),
    ("cpp_ss_ra",   "data/experiments/cpp/6_18_ecc7_fmf_standard_ss_ra_pubmed/output/output.aux"),
    ("py_ss_ra",    "data/experiments/python/6_23_ecc7_fmf_ss_ra_pubmed_python_abm/output/output.aux"),
    ("cpp_noss_sa", "data/experiments/cpp/6_19_ecc7_fmf_standard_no_ss_sa_pubmed/output/output.aux"),
    ("py_noss_sa",  "data/experiments/python/6_23_ecc7_fmf_no_ss_sa_pubmed_python_abm/output/output.aux"),
    ("cpp_ss_sa",   "data/experiments/cpp/6_19_ecc7_fmf_standard_ss_sa_pubmed/output/output.aux"),
    ("py_ss_sa",    "data/experiments/python/6_26_ecc7_fmf_ss_sa_pubmed_python_abm/output/output.aux")
]

# Build table
rows = []

for tag, file in experiments:
    df = pd.read_csv(file)
    df['in_degree'] = pd.to_numeric(df['in_degree'], errors='coerce')
    df = df[df['in_degree'].notnull() & (df['in_degree'] > 0)]

    for typ in ["seed", "agent"]:
        subset = df[df['type'] == typ]['in_degree']
        if subset.empty:
            continue
        stats = {
            "tag": tag,
            "type": typ,
            "min": subset.min(),
            "q1": round(subset.quantile(0.25)),
            "q2": round(subset.quantile(0.50)),
            "q3": round(subset.quantile(0.75)),
            "perc0.90": round(subset.quantile(0.90)),
            "perc0.99": round(subset.quantile(0.99), 2),
            "max": subset.max()
        }
        rows.append(stats)

# Make the final dataframe
summary_df = pd.DataFrame(rows, columns=["tag","type","min","q1","q2","q3","perc0.90","perc0.99","max"])

# Display it like your table
print(summary_df.to_string(index=False))
```
